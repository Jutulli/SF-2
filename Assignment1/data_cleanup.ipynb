```python
import pandas as pd
import matplotlib.pyplot as plit
import numpy as numpy
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
pd.options.display.max_rows = 1000
pd.options.display.max_columns = 100

data1 = pd.read_csv('Capstone_Data.csv')
#data1 = data1[5:len(data1)]

replaceIndex = 100000000
for index1, info in enumerate(data1):
    is_position = False

    if info[:8] == "Position":
        is_position = True
        replaceIndex = 100000000

    print(index1, replaceIndex)
        
    for index2, value in enumerate(data1[info]):
        if is_position and value <= 0.12 and replaceIndex == 100000000 :
            replaceIndex = index2


        if index2 >= replaceIndex:
            print(index2)
            data1[info][index2] = numpy.nan

data1 = data1.dropna(axis=0, how='all')
            
data1
        
```
